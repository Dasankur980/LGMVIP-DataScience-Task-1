```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
%matplotlib inline
```

```python
#Loading the columns into the Data by declaring the names of the columns
iris=pd.read_csv('iris.data',names=['Sepal Length','Sepal Width','Petal Length','Petal Width','Class'])
iris.head()
```

```python
#Dividing the Iris Information into x and y
x=iris.values[:,0:4]
y=iris.values[:,4]
```

```python
#Splicing the Iris Data into Train and Test set
from sklearn.model_selection import train_test_split
iris_x_train,iris_x_test,iris_y_train,iris_y_test=train_test_split(x,y,test_size=0.47,random_state=20)
print(iris_x_train)
```

### __Logistic Regression Model__

```python
#Importing Logistic Regression from sklearn.linearmodel and declaring it

from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(iris_x_train,iris_y_train)
```

```python
#Declaration of Predicting Variable

pred=lr.predict(iris_x_test)
```

```python
#F1 Score

f1=f1_score(iris_y_test,pred,average='micro')
print("F1 Score:- ",f1)
```

```python
#Accuracy Score

acuscore=accuracy_score(iris_y_test,pred)
print("Accuracy Score:- ",acuscore*100,"%")
```

### __Random Forest Classifier Model__

```python
#Importing Random Forest Classifier from sklearn.ensmeble and declaring it

from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(iris_x_train,iris_y_train)
```

```python
#Declaration of Predicting Variable

pred=rfc.predict(iris_x_test)
```

```python
#F1 Score

f1=f1_score(iris_y_test,pred,average='micro')
print("F1 Score:- ",f1)
```

```python
#Accuracy Score

acuscore=accuracy_score(iris_y_test,pred)
print("Accuracy Score:- ",acuscore*100,"%")
```

### __Support Vector Classification Model__

```python
#Importing SVC(Support Vector Classification) from sklearn.svm and declaring it

from sklearn.svm import SVC
svc=SVC()
svc.fit(iris_x_train,iris_y_train)
```

```python
#Declaration of Predicting Variable

pred=svc.predict(iris_x_test)
```

```python
#F1 Score

f1=f1_score(iris_y_test,pred,average='micro')
print("F1 Score:- ",f1)
```

```python
#Accuracy Score

acuscore=accuracy_score(iris_y_test,pred)
print("Accuracy Score:- ",acuscore*100,"%")
```

## __Confusion Matrix__

```python
print("Confusion Matrix:- \n",confusion_matrix(iris_y_test,pred))
```

## __Classification Report__

#### __Iris-Setosa:- 0.0__
#### __Iris-Versicolor:- 1.0__
#### __Iris-Virginica:- 2.0__

```python
print("Classification Report of Iris:- \n",classification_report(iris_y_test,pred))
```
